```python
# =============================================================================
# DOCUMENT FORGERY DETECTION – MODERATE TAMPERING (FORCE RETRAIN)
# =============================================================================
# Deletes any existing model and trains from scratch
# =============================================================================

# 1. Install dependencies
# =============================================================================
!pip install -q kagglehub scikit-learn opencv-python-headless

# 2. Imports
# =============================================================================
import os, random, numpy as np, pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image, ImageDraw, ImageFont
import cv2
from tqdm import tqdm
from sklearn.metrics import (
    accuracy_score, confusion_matrix, roc_curve, auc,
    precision_recall_curve, average_precision_score,
    mean_squared_error, roc_auc_score
)
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
import shutil
import pickle
import warnings
warnings.filterwarnings('ignore')

# -----------------------------------------------------------------------------
# CONFIG
# -----------------------------------------------------------------------------
class Config:
    DATA_DIR = 'data'
    MODELS_DIR = 'models'
    FIGURES_DIR = 'figures'
    EPOCHS = 25
    BATCH_SIZE = 32
    LR = 0.0005
    IMG_SIZE = 224
    NUM_SAMPLES = 4000
    TAMPER_PROB = 0.5
    DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

config = Config()
os.makedirs(config.DATA_DIR, exist_ok=True)
os.makedirs(config.MODELS_DIR, exist_ok=True)
os.makedirs(config.FIGURES_DIR, exist_ok=True)
print(f"🔧 Device: {config.DEVICE}")

# -----------------------------------------------------------------------------
# BALANCED TAMPERING – Moderate severity
# -----------------------------------------------------------------------------

def moderate_copy_move(image_np, severity=0.5):
    h, w = image_np.shape[:2]
    if h < 150 or w < 150:
        return image_np
    patch_h = random.randint(30, 50 + int(30*severity))
    patch_w = random.randint(30, 50 + int(30*severity))
    src_x = random.randint(0, w - patch_w - 10)
    src_y = random.randint(0, h - patch_h - 10)
    dst_x = random.randint(0, w - patch_w - 10)
    dst_y = random.randint(0, h - patch_h - 10)
    attempts = 0
    while abs(dst_x - src_x) < patch_w and abs(dst_y - src_y) < patch_h and attempts < 50:
        dst_x = random.randint(0, w - patch_w - 10)
        dst_y = random.randint(0, h - patch_h - 10)
        attempts += 1
    patch = image_np[src_y:src_y+patch_h, src_x:src_x+patch_w].copy()
    patch = cv2.GaussianBlur(patch, (3, 3), 0)
    alpha = random.uniform(0.6, 0.95)
    dst_region = image_np[dst_y:dst_y+patch_h, dst_x:dst_x+patch_w]
    blended = cv2.addWeighted(patch, alpha, dst_region, 1-alpha, 0)
    image_np[dst_y:dst_y+patch_h, dst_x:dst_x+patch_w] = blended
    return image_np

def moderate_splicing(image_np, donor_pool, severity=0.5):
    if not donor_pool or len(donor_pool) < 3:
        return image_np
    h, w = image_np.shape[:2]
    if h < 150 or w < 150:
        return image_np
    donor_path = random.choice(donor_pool)
    try:
        donor = cv2.imread(donor_path)
        if donor is None:
            return image_np
        donor = cv2.resize(donor, (w, h))
    except:
        return image_np
    patch_h = random.randint(30, 50 + int(20*severity))
    patch_w = random.randint(30, 50 + int(20*severity))
    src_x = random.randint(0, w - patch_w - 10)
    src_y = random.randint(0, h - patch_h - 10)
    patch = donor[src_y:src_y+patch_h, src_x:src_x+patch_w].copy()
    patch = cv2.GaussianBlur(patch, (3, 3), 0)
    dst_x = random.randint(0, w - patch_w - 10)
    dst_y = random.randint(0, h - patch_h - 10)
    alpha = random.uniform(0.5, 0.9)
    dst_region = image_np[dst_y:dst_y+patch_h, dst_x:dst_x+patch_w]
    blended = cv2.addWeighted(patch, alpha, dst_region, 1-alpha, 0)
    image_np[dst_y:dst_y+patch_h, dst_x:dst_x+patch_w] = blended
    return image_np

def moderate_text_replacement(image_np, severity=0.5):
    h, w = image_np.shape[:2]
    if h < 150 or w < 150:
        return image_np
    image_pil = Image.fromarray(cv2.cvtColor(image_np, cv2.COLOR_BGR2RGB))
    draw = ImageDraw.Draw(image_pil)
    rect_w = random.randint(50, 80 + int(30*severity))
    rect_h = random.randint(20, 30 + int(10*severity))
    x = random.randint(20, w - rect_w - 20)
    y = random.randint(20, h - rect_h - 20)
    bg_region = image_np[max(0,y-10):min(h,y+rect_h+10), max(0,x-10):min(w,x+rect_w+10)]
    if bg_region.size > 0:
        bg_color = tuple(np.mean(bg_region.reshape(-1,3), axis=0).astype(int).tolist())
    else:
        bg_color = (255,255,255)
    draw.rectangle([x, y, x+rect_w, y+rect_h], fill=tuple(bg_color))
    try:
        font = ImageFont.truetype("/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf", 14 + int(8*severity))
    except:
        font = ImageFont.load_default()
    new_text = ''.join(random.choices('0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ', k=random.randint(5, 8)))
    text_color = (0, 0, 0) if random.random() > 0.3 else (30, 30, 30)
    draw.text((x+3, y+2), new_text, fill=text_color, font=font)
    image_np = cv2.cvtColor(np.array(image_pil), cv2.COLOR_RGB2BGR)
    return image_np

def apply_moderate_tampering(image_path, donor_pool=None):
    img = cv2.imread(image_path)
    if img is None:
        return None, None
    severity = random.uniform(0.3, 0.7)
    tamper_type = random.choice(['copy_move', 'splicing', 'text_replacement'])
    if tamper_type == 'copy_move':
        tampered = moderate_copy_move(img, severity)
    elif tamper_type == 'splicing':
        tampered = moderate_splicing(img, donor_pool, severity)
    else:
        tampered = moderate_text_replacement(img, severity)
    return tampered, tamper_type

# -----------------------------------------------------------------------------
# GENERATE DATASET
# -----------------------------------------------------------------------------
def generate_dataset():
    import kagglehub
    output_dir = os.path.join(config.DATA_DIR, 'moderate_tamper')
    authentic_dir = os.path.join(output_dir, 'authentic')
    tampered_dir = os.path.join(output_dir, 'tampered')
    os.makedirs(authentic_dir, exist_ok=True)
    os.makedirs(tampered_dir, exist_ok=True)
    if len(os.listdir(authentic_dir)) > 500 and len(os.listdir(tampered_dir)) > 500:
        print(f"✅ Moderate tamper dataset already exists: {output_dir}")
        return output_dir
    print("📥 Downloading EACS dataset...")
    eacs_path = kagglehub.dataset_download("baizidmdashadzzaman/exam-admit-card-dataset")
    image_files = []
    for root, _, files in os.walk(eacs_path):
        for f in files:
            if f.lower().endswith(('.jpg', '.jpeg', '.png')):
                image_files.append(os.path.join(root, f))
    print(f"Found {len(image_files)} images in EACS.")
    random.shuffle(image_files)
    num_samples = min(config.NUM_SAMPLES, len(image_files))
    authentic_images = image_files[:num_samples//2]
    tamper_sources = image_files[num_samples//2:num_samples]
    print(f"Creating {len(authentic_images)} authentic and {len(tamper_sources)} tampered images...")
    for i, src in enumerate(tqdm(authentic_images, desc="Authentic")):
        img = cv2.imread(src)
        if img is None:
            continue
        if random.random() < 0.3:
            img = cv2.convertScaleAbs(img, alpha=1+random.uniform(-0.1,0.1), beta=random.randint(-15,15))
        cv2.imwrite(os.path.join(authentic_dir, f"{i}.jpg"), img)
    for i, src in enumerate(tqdm(tamper_sources, desc="Tampered")):
        donor_pool = random.sample(authentic_images, min(30, len(authentic_images)))
        result, tamper_type = apply_moderate_tampering(src, donor_pool)
        if result is not None:
            cv2.imwrite(os.path.join(tampered_dir, f"{i}_{tamper_type}.jpg"), result)
    print(f"✅ Moderate tamper dataset created at {output_dir}")
    return output_dir

# -----------------------------------------------------------------------------
# DATASET CLASS
# -----------------------------------------------------------------------------
class UnifiedDocumentDataset(Dataset):
    def __init__(self, data_root, transform=None):
        self.data_root = data_root
        self.transform = transform
        self.samples = []
        authentic_dir = os.path.join(data_root, 'authentic')
        tampered_dir = os.path.join(data_root, 'tampered')
        if os.path.exists(authentic_dir):
            for f in os.listdir(authentic_dir):
                if f.lower().endswith(('.jpg', '.jpeg', '.png')):
                    self.samples.append((os.path.join(authentic_dir, f), 0))
        if os.path.exists(tampered_dir):
            for f in os.listdir(tampered_dir):
                if f.lower().endswith(('.jpg', '.jpeg', '.png')):
                    self.samples.append((os.path.join(tampered_dir, f), 1))
        self._balance_samples()
        print(f"[Dataset] Total: {len(self.samples)} (Genuine: {sum(1 for _,l in self.samples if l==0)}, Tampered: {sum(1 for _,l in self.samples if l==1)})")
        if self.transform is None:
            self.transform = self._default_transform()

    def _balance_samples(self):
        genuine = [(p,l) for p,l in self.samples if l==0]
        tampered = [(p,l) for p,l in self.samples if l==1]
        target = max(len(genuine), len(tampered))
        if len(genuine) < target:
            genuine = genuine * (target // len(genuine) + 1)
            genuine = genuine[:target]
        if len(tampered) < target:
            tampered = tampered * (target // len(tampered) + 1)
            tampered = tampered[:target]
        self.samples = genuine + tampered
        random.shuffle(self.samples)

    def _default_transform(self):
        return transforms.Compose([
            transforms.Resize((config.IMG_SIZE, config.IMG_SIZE)),
            transforms.RandomHorizontalFlip(),
            transforms.RandomRotation(10),
            transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                 std=[0.229, 0.224, 0.225])
        ])

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, label = self.samples[idx]
        try:
            image = Image.open(img_path).convert('RGB')
        except:
            image = Image.new('RGB', (config.IMG_SIZE, config.IMG_SIZE), color='gray')
        if self.transform:
            image = self.transform(image)
        return image, label

# -----------------------------------------------------------------------------
# MODEL
# -----------------------------------------------------------------------------
class UnifiedDocumentModel(nn.Module):
    def __init__(self, num_classes=2):
        super(UnifiedDocumentModel, self).__init__()
        self.backbone = models.resnet50(pretrained=True)
        self.backbone.fc = nn.Linear(2048, num_classes)

    def forward(self, x):
        return self.backbone(x)

# -----------------------------------------------------------------------------
# TRAINING
# -----------------------------------------------------------------------------
def train_model(data_root):
    print("\n" + "="*60)
    print("TRAINING UNIFIED MODEL")
    print("="*60)

    dataset = UnifiedDocumentDataset(data_root)
    if len(dataset) == 0:
        print("❌ No samples found!")
        return None, None

    labels = [label for _, label in dataset.samples]
    train_idx, val_idx = train_test_split(
        range(len(dataset)),
        test_size=0.2,
        stratify=labels,
        random_state=42
    )

    train_transform = transforms.Compose([
        transforms.Resize((config.IMG_SIZE, config.IMG_SIZE)),
        transforms.RandomHorizontalFlip(),
        transforms.RandomRotation(10),
        transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                             std=[0.229, 0.224, 0.225])
    ])
    val_transform = transforms.Compose([
        transforms.Resize((config.IMG_SIZE, config.IMG_SIZE)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                             std=[0.229, 0.224, 0.225])
    ])

    class TransformSubset(torch.utils.data.Dataset):
        def __init__(self, subset, transform):
            self.subset = subset
            self.transform = transform
        def __len__(self):
            return len(self.subset)
        def __getitem__(self, idx):
            img_path, label = dataset.samples[self.subset.indices[idx]]
            try:
                image = Image.open(img_path).convert('RGB')
            except:
                image = Image.new('RGB', (config.IMG_SIZE, config.IMG_SIZE), color='gray')
            if self.transform:
                image = self.transform(image)
            return image, label

    train_ds = TransformSubset(torch.utils.data.Subset(dataset, train_idx), train_transform)
    val_ds = TransformSubset(torch.utils.data.Subset(dataset, val_idx), val_transform)

    train_loader = DataLoader(train_ds, batch_size=config.BATCH_SIZE, shuffle=True)
    val_loader = DataLoader(val_ds, batch_size=config.BATCH_SIZE, shuffle=False)

    print(f"Train: {len(train_ds)}, Val: {len(val_ds)}")

    model = UnifiedDocumentModel().to(config.DEVICE)
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.AdamW(model.parameters(), lr=config.LR, weight_decay=1e-4)
    scheduler = optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=config.EPOCHS)

    best_acc = 0.0
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}

    for epoch in range(config.EPOCHS):
        model.train()
        train_loss, train_correct = 0.0, 0
        for images, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{config.EPOCHS}"):
            images, labels = images.to(config.DEVICE), labels.to(config.DEVICE)
            optimizer.zero_grad()
            outputs = model(images)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()
            train_correct += (outputs.argmax(1) == labels).sum().item()

        model.eval()
        val_loss, val_correct = 0.0, 0
        with torch.no_grad():
            for images, labels in val_loader:
                images, labels = images.to(config.DEVICE), labels.to(config.DEVICE)
                outputs = model(images)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                val_correct += (outputs.argmax(1) == labels).sum().item()

        train_acc = train_correct / len(train_ds)
        val_acc = val_correct / len(val_ds)

        history['train_loss'].append(train_loss / len(train_loader))
        history['val_loss'].append(val_loss / len(val_loader))
        history['train_acc'].append(train_acc)
        history['val_acc'].append(val_acc)

        scheduler.step()

        print(f"Epoch {epoch+1}: Train Acc={train_acc:.4f}, Val Acc={val_acc:.4f}")
        if val_acc > best_acc:
            best_acc = val_acc
            torch.save(model.state_dict(), os.path.join(config.MODELS_DIR, 'unified_best.pth'))
            print(f"  ✅ Saved best (val_acc={best_acc:.4f})")

    with open(os.path.join(config.MODELS_DIR, 'history.pkl'), 'wb') as f:
        pickle.dump(history, f)

    return model, history

# -----------------------------------------------------------------------------
# EVALUATION
# -----------------------------------------------------------------------------
def evaluate_and_plot(model_path, data_root):
    print("\n" + "="*60)
    print("EVALUATING MODEL")
    print("="*60)

    model = UnifiedDocumentModel().to(config.DEVICE)
    if not os.path.exists(model_path):
        print("❌ Model not found.")
        return
    model.load_state_dict(torch.load(model_path, map_location=config.DEVICE))
    model.eval()
    print("✅ Model loaded.")

    dataset = UnifiedDocumentDataset(data_root, transform=None)
    labels = [label for _, label in dataset.samples]
    _, val_idx = train_test_split(
        range(len(dataset)),
        test_size=0.2,
        stratify=labels,
        random_state=42
    )
    val_transform = transforms.Compose([
        transforms.Resize((config.IMG_SIZE, config.IMG_SIZE)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406],
                             std=[0.229, 0.224, 0.225])
    ])

    class EvalSubset(torch.utils.data.Dataset):
        def __init__(self, subset, transform):
            self.subset = subset
            self.transform = transform
        def __len__(self):
            return len(self.subset)
        def __getitem__(self, idx):
            img_path, label = dataset.samples[self.subset.indices[idx]]
            try:
                image = Image.open(img_path).convert('RGB')
            except:
                image = Image.new('RGB', (config.IMG_SIZE, config.IMG_SIZE), color='gray')
            if self.transform:
                image = self.transform(image)
            return image, label

    val_ds = EvalSubset(torch.utils.data.Subset(dataset, val_idx), val_transform)
    val_loader = DataLoader(val_ds, batch_size=config.BATCH_SIZE, shuffle=False)

    all_labels, all_preds, all_probs = [], [], []
    with torch.no_grad():
        for images, labels in val_loader:
            images = images.to(config.DEVICE)
            outputs = model(images)
            probs = torch.softmax(outputs, dim=1)
            preds = torch.argmax(outputs, dim=1)
            all_labels.extend(labels.cpu().numpy())
            all_preds.extend(preds.cpu().numpy())
            all_probs.extend(probs[:,1].cpu().numpy())

    all_labels = np.array(all_labels)
    all_preds = np.array(all_preds)
    all_probs = np.array(all_probs)

    acc = accuracy_score(all_labels, all_preds)
    tn, fp, fn, tp = confusion_matrix(all_labels, all_preds).ravel()
    sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0.0
    specificity = tn / (tn + fp) if (tn + fp) > 0 else 0.0
    rmse = np.sqrt(mean_squared_error(all_labels, all_probs))
    roc_auc = roc_auc_score(all_labels, all_probs)

    print("\n" + "="*60)
    print("MODEL PERFORMANCE METRICS")
    print("="*60)
    print(f"{'Metric':<20} {'Value':>10}")
    print("-"*60)
    print(f"{'Accuracy':<20} {acc*100:>9.2f}%")
    print(f"{'Sensitivity':<20} {sensitivity*100:>9.2f}%")
    print(f"{'Specificity':<20} {specificity*100:>9.2f}%")
    print(f"{'RMSE':<20} {rmse:>10.4f}")
    print(f"{'ROC-AUC':<20} {roc_auc:>10.4f}")
    print("="*60)
    print(f"\nConfusion Matrix: TN={tn}, FP={fp}, FN={fn}, TP={tp}")

    # ---- Plots ----
    cm = confusion_matrix(all_labels, all_preds)
    cm_pct = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis] * 100
    annot = np.empty_like(cm, dtype='object')
    for i in range(2):
        for j in range(2):
            annot[i,j] = f"{cm[i,j]}\n({cm_pct[i,j]:.1f}%)"

    fig, ax = plt.subplots(figsize=(6,5))
    sns.heatmap(cm, annot=annot, fmt='', cmap='Blues',
                xticklabels=['Genuine', 'Tampered'],
                yticklabels=['Genuine', 'Tampered'],
                annot_kws={'size':14, 'weight':'bold'})
    ax.set_title('Confusion Matrix')
    plt.tight_layout()
    plt.savefig(os.path.join(config.FIGURES_DIR, 'confusion_matrix.png'), dpi=300)
    plt.show()

    fpr, tpr, _ = roc_curve(all_labels, all_probs)
    roc_auc_val = auc(fpr, tpr)
    fig, ax = plt.subplots(figsize=(7,6))
    ax.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC (AUC = {roc_auc_val:.3f})')
    ax.plot([0,1], [0,1], color='navy', lw=2, linestyle='--', label='Random')
    ax.set_xlim([0.0, 1.0]); ax.set_ylim([0.0, 1.05])
    ax.set_xlabel('False Positive Rate')
    ax.set_ylabel('True Positive Rate')
    ax.set_title('ROC Curve')
    ax.legend(loc='lower right')
    ax.grid(True)
    plt.tight_layout()
    plt.savefig(os.path.join(config.FIGURES_DIR, 'roc_curve.png'), dpi=300)
    plt.show()

    history_path = os.path.join(config.MODELS_DIR, 'history.pkl')
    if os.path.exists(history_path):
        with open(history_path, 'rb') as f:
            history = pickle.load(f)
        fig, axes = plt.subplots(1, 2, figsize=(14, 5))
        epochs = range(1, len(history['train_loss'])+1)
        axes[0].plot(epochs, history['train_loss'], 'b-', label='Train Loss')
        axes[0].plot(epochs, history['val_loss'], 'r-', label='Val Loss')
        axes[0].set_xlabel('Epoch'); axes[0].set_ylabel('Loss')
        axes[0].set_title('Loss Curves'); axes[0].legend(); axes[0].grid(True)
        axes[1].plot(epochs, history['train_acc'], 'b-', label='Train Acc')
        axes[1].plot(epochs, history['val_acc'], 'r-', label='Val Acc')
        axes[1].set_xlabel('Epoch'); axes[1].set_ylabel('Accuracy')
        axes[1].set_title('Accuracy Curves'); axes[1].legend(); axes[1].grid(True)
        plt.tight_layout()
        plt.savefig(os.path.join(config.FIGURES_DIR, 'training_history.png'), dpi=300)
        plt.show()

    print(f"\n📁 All figures saved in: {config.FIGURES_DIR}")

# -----------------------------------------------------------------------------
# MAIN – FORCES RETRAINING BY DELETING OLD MODEL
# -----------------------------------------------------------------------------
def main():
    print("="*60)
    print("DOCUMENT FORGERY DETECTION – MODERATE TAMPERING (FORCE RETRAIN)")
    print("="*60)

    # Delete any existing model to force retraining
    model_path = os.path.join(config.MODELS_DIR, 'unified_best.pth')
    if os.path.exists(model_path):
        os.remove(model_path)
        print("🗑️ Deleted old model file. Training from scratch...")

    # Generate dataset (if not already present)
    data_root = generate_dataset()

    # Train the model
    train_model(data_root)

    # Evaluate and plot
    evaluate_and_plot(model_path, data_root)

    print("\n" + "="*60)
    print("✅ ALL TASKS COMPLETED!")
    print("="*60)
    print(f"\n📁 Figures saved in: {config.FIGURES_DIR}")

if __name__ == "__main__":
    main()
```

```python
# =============================================================================
# BAR CHART FOR PERFORMANCE METRICS (FIXED PATH)
# =============================================================================
# Automatically finds the dataset subfolder (moderate_tamper, subtle_tamper, etc.)
# =============================================================================

import os, random, numpy as np, matplotlib.pyplot as plt
import torch
from sklearn.metrics import (
    accuracy_score, confusion_matrix, roc_auc_score,
    mean_squared_error
)
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader
from torchvision import transforms
from PIL import Image
import torch.nn as nn
from torchvision import models
import warnings
warnings.filterwarnings('ignore')

# -----------------------------------------------------------------------------
# CONFIG
# -----------------------------------------------------------------------------
class Config:
    DATA_DIR = 'data'
    MODELS_DIR = 'models'
    FIGURES_DIR = 'figures'
    IMG_SIZE = 224
    BATCH_SIZE = 32
    DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

config = Config()
os.makedirs(config.FIGURES_DIR, exist_ok=True)

# -----------------------------------------------------------------------------
# FIND DATASET SUBFOLDER
# -----------------------------------------------------------------------------
def find_dataset_folder():
    """Look for a subfolder in data/ that contains 'authentic' and 'tampered'."""
    for root, dirs, files in os.walk(config.DATA_DIR):
        if 'authentic' in dirs and 'tampered' in dirs:
            # Check if there are images
            auth_path = os.path.join(root, 'authentic')
            tam_path = os.path.join(root, 'tampered')
            if len(os.listdir(auth_path)) > 0 and len(os.listdir(tam_path)) > 0:
                return root
    # If not found, use the last generated folder
    possible_folders = ['moderate_tamper', 'subtle_tamper', 'synthetic_dataset']
    for folder in possible_folders:
        path = os.path.join(config.DATA_DIR, folder)
        if os.path.exists(path):
            return path
    return None

data_root = find_dataset_folder()
if data_root is None:
    raise FileNotFoundError("Could not find dataset folder with 'authentic' and 'tampered' subfolders.")
print(f"📂 Using dataset: {data_root}")

# -----------------------------------------------------------------------------
# RE-DEFINE DATASET & MODEL (same as training)
# -----------------------------------------------------------------------------
class UnifiedDocumentDataset:
    def __init__(self, data_root, transform=None):
        self.data_root = data_root
        self.transform = transform
        self.samples = []
        authentic_dir = os.path.join(data_root, 'authentic')
        tampered_dir = os.path.join(data_root, 'tampered')
        if os.path.exists(authentic_dir):
            for f in os.listdir(authentic_dir):
                if f.lower().endswith(('.jpg', '.jpeg', '.png')):
                    self.samples.append((os.path.join(authentic_dir, f), 0))
        if os.path.exists(tampered_dir):
            for f in os.listdir(tampered_dir):
                if f.lower().endswith(('.jpg', '.jpeg', '.png')):
                    self.samples.append((os.path.join(tampered_dir, f), 1))
        # Balance classes
        genuine = [(p,l) for p,l in self.samples if l==0]
        tampered = [(p,l) for p,l in self.samples if l==1]
        target = max(len(genuine), len(tampered))
        if len(genuine) < target:
            genuine = genuine * (target // len(genuine) + 1)
            genuine = genuine[:target]
        if len(tampered) < target:
            tampered = tampered * (target // len(tampered) + 1)
            tampered = tampered[:target]
        self.samples = genuine + tampered
        random.shuffle(self.samples)
        print(f"[Dataset] Loaded {len(self.samples)} samples (Genuine: {sum(1 for _,l in self.samples if l==0)}, Tampered: {sum(1 for _,l in self.samples if l==1)})")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, label = self.samples[idx]
        try:
            image = Image.open(img_path).convert('RGB')
        except:
            image = Image.new('RGB', (config.IMG_SIZE, config.IMG_SIZE), color='gray')
        if self.transform:
            image = self.transform(image)
        return image, label

class UnifiedDocumentModel(nn.Module):
    def __init__(self, num_classes=2):
        super(UnifiedDocumentModel, self).__init__()
        self.backbone = models.resnet50(pretrained=False)
        self.backbone.fc = nn.Linear(2048, num_classes)

    def forward(self, x):
        return self.backbone(x)

# -----------------------------------------------------------------------------
# LOAD MODEL
# -----------------------------------------------------------------------------
model_path = os.path.join(config.MODELS_DIR, 'unified_best.pth')
if not os.path.exists(model_path):
    raise FileNotFoundError(f"Model not found at {model_path}")

model = UnifiedDocumentModel().to(config.DEVICE)
model.load_state_dict(torch.load(model_path, map_location=config.DEVICE))
model.eval()
print("✅ Model loaded.")

# -----------------------------------------------------------------------------
# LOAD DATASET & CREATE VALIDATION LOADER
# -----------------------------------------------------------------------------
dataset = UnifiedDocumentDataset(data_root, transform=None)
if len(dataset) == 0:
    raise ValueError("Dataset is empty. Check the dataset folder.")
labels = [label for _, label in dataset.samples]
_, val_idx = train_test_split(
    range(len(dataset)),
    test_size=0.2,
    stratify=labels,
    random_state=42
)

val_transform = transforms.Compose([
    transforms.Resize((config.IMG_SIZE, config.IMG_SIZE)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                         std=[0.229, 0.224, 0.225])
])

class EvalSubset(torch.utils.data.Dataset):
    def __init__(self, subset, transform):
        self.subset = subset
        self.transform = transform
    def __len__(self):
        return len(self.subset)
    def __getitem__(self, idx):
        img_path, label = dataset.samples[self.subset.indices[idx]]
        try:
            image = Image.open(img_path).convert('RGB')
        except:
            image = Image.new('RGB', (config.IMG_SIZE, config.IMG_SIZE), color='gray')
        if self.transform:
            image = self.transform(image)
        return image, label

val_ds = EvalSubset(torch.utils.data.Subset(dataset, val_idx), val_transform)
val_loader = DataLoader(val_ds, batch_size=config.BATCH_SIZE, shuffle=False)

# -----------------------------------------------------------------------------
# EVALUATE
# -----------------------------------------------------------------------------
all_labels, all_preds, all_probs = [], [], []
with torch.no_grad():
    for images, labels in val_loader:
        images = images.to(config.DEVICE)
        outputs = model(images)
        probs = torch.softmax(outputs, dim=1)
        preds = torch.argmax(outputs, dim=1)
        all_labels.extend(labels.cpu().numpy())
        all_preds.extend(preds.cpu().numpy())
        all_probs.extend(probs[:,1].cpu().numpy())

all_labels = np.array(all_labels)
all_preds = np.array(all_preds)
all_probs = np.array(all_probs)

# ---- Compute metrics ----
acc = accuracy_score(all_labels, all_preds)
tn, fp, fn, tp = confusion_matrix(all_labels, all_preds).ravel()
sensitivity = tp / (tp + fn) if (tp + fn) > 0 else 0.0
specificity = tn / (tn + fp) if (tn + fp) > 0 else 0.0
rmse = np.sqrt(mean_squared_error(all_labels, all_probs))
roc_auc = roc_auc_score(all_labels, all_probs)

print("\n" + "="*60)
print("COMPUTED METRICS:")
print("="*60)
print(f"Accuracy:     {acc*100:.2f}%")
print(f"Sensitivity:  {sensitivity*100:.2f}%")
print(f"Specificity:  {specificity*100:.2f}%")
print(f"RMSE:         {rmse:.4f}")
print(f"ROC-AUC:      {roc_auc:.4f}")
print("="*60)

# -----------------------------------------------------------------------------
# BAR CHART
# -----------------------------------------------------------------------------
metrics = {
    'Accuracy': acc*100,
    'Sensitivity': sensitivity*100,
    'Specificity': specificity*100,
    'ROC-AUC': roc_auc*100,
    'RMSE': rmse * 100   # scale RMSE to percentages for visual comparison
}

names = list(metrics.keys())
values = list(metrics.values())
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd']

plt.figure(figsize=(9,6))
bars = plt.bar(names, values, color=colors, edgecolor='black', linewidth=1.2)

# Add value labels
for bar, val in zip(bars, values):
    plt.text(bar.get_x() + bar.get_width()/2., val + 0.5,
             f'{val:.1f}%', ha='center', va='bottom', fontsize=11, fontweight='bold')

plt.ylabel('Score (%)', fontsize=14)
plt.title('Model Performance Metrics', fontsize=16, fontweight='bold')
plt.ylim(0, max(values) + 8)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Save figure
save_path = os.path.join(config.FIGURES_DIR, 'performance_bar_chart.png')
plt.savefig(save_path, dpi=300, bbox_inches='tight')
print(f"\n✅ Bar chart saved to: {save_path}")
plt.show()
```
